# Preprocess the sleep column

```python
df1['Sleep Duration'] = df1['Sleep Duration'].str.strip("'").str.strip()
print(df1["Sleep Duration"].unique())
sleep_map = {
    'Less than 5 hours': 4,
    '5-6 hours': 5.5,
    '7-8 hours': 7.5,
    'More than 8 hours': 9,
    'Others': (4+5.5+7.5+9)/4  
}

df1['SleepHours'] = df1['Sleep Duration'].map(sleep_map)
```

```python
df1[df1["Sleep Duration"]=="Others"]
```

```python
print(df1['Dietary Habits'].unique())
diet_map = {
    'Healthy': 1.,
    'Moderate': 0.5,
    'Unhealthy': 0.,
    'Others': 0.5  
}
```

```python
df1['Dietary Habits'] = df1['Dietary Habits'].map(diet_map)
```

```python
df1["Have you ever had suicidal thoughts ?"]=df1["Have you ever had suicidal thoughts ?"].replace("No",'0')
df1["Have you ever had suicidal thoughts ?"]=df1["Have you ever had suicidal thoughts ?"].replace("Yes",'1')
```

# Finding Outiers and removing

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=data)
plt.show()
```

```python
def remove_outliers(df,column):
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    low=q1-1.5*iqr
    high=q3+1.5*iqr
    df=df[(df[column]>=low) & (df[column]<=high)]
    return df
for i in ["Age","CGPA"]:
    data=remove_outliers(data,i)
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(data=data)
plt.show()
```
